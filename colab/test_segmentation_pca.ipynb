```
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
x1 = x1.reshape((36*36, 1024))
```

```
x1 = StandardScaler().fit_transform(x1)
pca = PCA(n_components=3)
x2 = pca.fit_transform(x1)
x2.shape
```

```
x3 = x2.copy()
```

```
x3[:,0] = (x3[:,0] - x3[:,0].min())/(x3[:,0].max() - x3[:,0].min())
x3[:,1] = (x3[:,1] - x3[:,1].min())/(x3[:,1].max() - x3[:,1].min())
x3[:,2] = (x3[:,2] - x3[:,2].min())/(x3[:,2].max() - x3[:,2].min())
```

```
x4 = x3.copy()
x4 = x4.reshape((36,36,3))
from PIL import Image
im = Image.fromarray(np.uint8(x4*255.0), 'RGB')
new_img = im.resize((590,428),Image.BICUBIC)
im.show()
new_img.save('/content/drive/My Drive/Segmentation/image.jpg')
```
